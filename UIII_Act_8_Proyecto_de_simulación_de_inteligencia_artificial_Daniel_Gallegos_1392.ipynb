```python
# Simulador de Inteligencia Artificial: Arquitecto Virtual
# Nivel: Principiante (Sin funciones, sin POO, sin ciclos for)

print("==================================================")
print(" ¡Bienvenido al Diseñador de Planos IA v1.0!      ")
print(" Tu asistente virtual de arquitectura e ingeniería")
print("==================================================")

# Inicializamos variables de control
continuar_programa = "si"

while continuar_programa == "si" or continuar_programa == "SI" or continuar_programa == "Si":
    print("\n--- MENÚ PRINCIPAL DEL PROYECTO ---")
    print("1. Definir terreno y tamaño de la casa")
    print("2. Agregar habitaciones y estilo")
    print("3. Calcular presupuesto estimado")
    print("4. Mostrar el plano final generado por la IA")
    print("5. Salir")

    opcion = input("\nSelecciona una opción (1-5): ")

    # -------------------------------------------------------------
    # OPCIÓN 1: DEFINIR TERRENO
    # -------------------------------------------------------------
    if opcion == "1":
        print("\n[IA] Analizando dimensiones...")
        metros_frente = input("¿Cuántos metros de frente tiene tu terreno?: ")
        metros_fondo = input("¿Cuántos metros de fondo tiene tu terreno?: ")

        # Conversión simple de texto a número entero
        frente = int(metros_frente)
        fondo = int(metros_fondo)
        area_total = frente * fondo

        print("\n> ¡Análisis listo!")
        print("El área total de tu terreno es de", area_total, "metros cuadrados.")

        if area_total < 50:
            print("[IA Consejo]: Tu terreno es pequeño. Recomiendo un plano de 2 pisos para optimizar.")
            pisos = "2 pisos"
        elif area_total >= 50 and area_total <= 150:
            print("[IA Consejo]: Tamaño ideal para una casa residencial de 1 o 2 pisos.")
            pisos = "1 piso espacioso"
        else:
            print("[IA Consejo]: ¡Terreno enorme! Tienes espacio para jardín, alberca y cochera triple.")
            pisos = "Residencia de lujo"

    # -------------------------------------------------------------
    # OPCIÓN 2: AGREGAR HABITACIONES Y ESTILO
    # -------------------------------------------------------------
    elif opcion == "2":
        print("\n--- CONFIGURADOR DE ESPACIOS ---")
        num_cuartos = input("¿Cuántas recámaras deseas incluir en el plano?: ")
        num_banos = input("¿Cuántos baños completos?: ")

        print("\nSelecciona el estilo arquitectónico:")
        print("- Moderno (Líneas rectas, grandes ventanas)")
        print("- Rústico (Madera, piedra, calidez)")
        print("- Minimalista (Espacios abiertos, pocos muros)")
        estilo = input("Escribe el estilo que prefieras: ")

        print("\n> [IA] Distribución espacial guardada con éxito.")

    # -------------------------------------------------------------
    # OPCIÓN 3: CALCULAR PRESUPUESTO
    # -------------------------------------------------------------
    elif opcion == "3":
        print("\n--- CALCULADORA DE COSTOS IA ---")
        print("Elige la calidad de los acabados para los planos:")
        print("1. Económico")
        print("2. Estándar")
        print("3. Premium")

        calidad = input("Selecciona (1, 2 o 3): ")

        if calidad == "1":
            costo_metro = 8000
            tipo_acabado = "Económico"
        elif calidad == "2":
            costo_metro = 12000
            tipo_acabado = "Estándar"
        elif calidad == "3":
            costo_metro = 20000
            tipo_acabado = "Premium"
        else:
            costo_metro = 10000
            tipo_acabado = "No definido (por defecto Estándar)"

        # Para el cálculo asumimos una casa base de 90 metros cuadrados si no han ido a la opción 1
        costo_aproximado = 90 * costo_metro
        print("\n> [IA Reporte]: Con acabados", tipo_acabado, "el costo estimado de construcción")
        print("para una casa promedio es de: $", costo_aproximado, "MXN.")

    # -------------------------------------------------------------
    # OPCIÓN 4: MOSTRAR PLANO FINAL (SIMULACIÓN VISUAL)
    # -------------------------------------------------------------
    elif opcion == "4":
        print("\n==================================================")
        print("      PLANO GENERADO POR LA INTELIGENCIA ARTIFICIAL")
        print("==================================================")
        print(" [Fachada Principal] -> Estilo:", estilo)
        print(" [Planta Baja]       -> Cochera | Sala | Comedor | Cocina")
        print(" [Planta Alta]       ->", num_cuartos, "Recámaras |", num_banos, "Baños")
        print(" [Estructura]        -> Recomendación:", pisos)
        print("==================================================")
        print("           _________________________")
        print("          /                        \\")
        print("         /   [Recámara 1] [Baño]    \\")
        print("        /____________________________\\")
        print("        |                            |")
        print("        |   [  Sala  ]  [ Cocina ]   |")
        print("        |____________________________|")
        print("\n ¡Tu plano digital ha sido renderizado con éxito!")

    # -------------------------------------------------------------
    # OPCIÓN 5: SALIR
    # -------------------------------------------------------------
    elif opcion == "5":
        print("\n¡Gracias por diseñar con la IA de Arquitectura! Hasta luego.")
        continuar_programa = "no" # Rompe el ciclo while

    else:
        print("\n[Error] Opción no válida. Intenta de nuevo con números del 1 al 5.")

    # Pregunta de control para mantener el menú vivo si no eligió salir directamente
    if opcion != "5":
        print("\n--------------------------------------------------")
        continuar_programa = input("¿Deseas volver al menú principal? (si/no): ")

print("\n=== Fin de la simulación ===")
```
